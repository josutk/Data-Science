```python
import pandas as pd

pd.read_csv("employee_reviews.csv", encoding="utf-8") 
dataframe = pd.read_csv("employee_reviews.csv");
```

```python
import pandas as pd
import nltk as nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from textblob import TextBlob


dataframe2 = dataframe[dataframe.columns[6:8]]
dataframe2[dataframe.columns[1]] = dataframe.columns[1]
dataframe2[dataframe.columns[1]] = dataframe[dataframe.columns[1]].values

nltk.download('stopwords')
nltk.download('punkt')
stop = set(stopwords.words('english'))

dataframe2['pros'] = dataframe2['pros'].apply(word_tokenize)
dataframe2['pros'] = dataframe2['pros'].apply(lambda x: [item for item in x if item not in stop])


dataframe2['cons'] = dataframe2['cons'].apply(word_tokenize)
dataframe2['cons'] = dataframe2['cons'].apply(lambda x: [item for item in x if item not in stop])

dataframe2['cons'] = dataframe2['cons'].apply(' '.join)
dataframe2['pros'] = dataframe2['pros'].apply(' '.join)
dataframe2.head()
```

```python
def get_sentiment_analysis(column):
    polarity = []
    subjectivity =[]
    for item in column:
        text = TextBlob(item)
        polarity.append(text.sentiment.polarity)
        subjectivity.append(text.sentiment.subjectivity)
    return list(zip(polarity, subjectivity))



sementiments_pros = get_sentiment_analysis(dataframe2['pros']) 
sentimentsframe_pros = pd.DataFrame(sementiments_pros, columns = ['pros_polarity', 'pros_subjective'])

sementiments_cons = get_sentiment_analysis(dataframe2['cons']) 
sentimentsframe_cons = pd.DataFrame(sementiments_cons, columns = ['cons_polarity', 'cons_subjective'])
sentimentsframe.head()
```
